```python
# Task 2 - Marks Analysis Using Loops & Conditionals

student_name = "Ayesha Sharma"
subjects     = ["Math", "Physics", "CS", "English", "Chemistry"]
marks        = [88, 72, 95, 60, 78]

# 1. Using a for loop, print each subject alongside its marks and a grade label based on this scheme.
for i in range(len(subjects)):
    subject = subjects[i]
    mark = marks[i]
    if mark >= 90:
        grade = "A+"
    elif mark >= 80:
        grade = "A" 
    elif mark >= 70:
        grade = "B"
    elif mark >= 60:
        grade = "C"
    else:
        grade = "F"
    print(f"{subject}: {mark} - Grade: {grade}")

# 2. Calculations and Print:

print () # Add an extra newline for better separation between sections
total_marks = sum(marks)
print(f"Total marks: {total_marks}")
average_marks = total_marks / len(marks)
print(f"Average marks: {round(average_marks, 2)}")
highest_mark = max(marks)
highest_subject = subjects[marks.index(highest_mark)]
print(f"Highest scoring subject: {highest_subject} - {highest_mark}")
lowest_mark = min(marks)
lowest_subject = subjects[marks.index(lowest_mark)]
print(f"Lowest scoring subject: {lowest_subject} - {lowest_mark}")

# 3. Using a while loop, simulate a marks-entry system that allows adding new subjects:
# In each iteration, first ask for a subject name, then ask for the marks (0–100) for that subject.
# Stop the loop when the user types done as the subject name.

new_subjects_count = 0

while True:
    subject_name = input("Enter subject name (or 'done' to stop): ").strip()
    
    if subject_name.lower() == "done":
        break
    
    marks_input = input(f"Enter marks for {subject_name} (0-100): ").strip()
    
    # Validate: must be numeric
    if not marks_input.replace('.', '', 1).isdigit():
        print("⚠️ Warning: Invalid input. Marks must be a number. Entry not added.\n")
        continue
    
    marks_value = float(marks_input)
    
    # Validate: must be in range 0-100
    if marks_value < 0 or marks_value > 100:
        print("⚠️ Warning: Marks must be between 0 and 100. Entry not added.\n")
        continue
    
    # Valid entry — add to lists
    subjects.append(subject_name)
    marks.append(marks_value)
    new_subjects_count += 1
    print(f"✅ '{subject_name}' with marks {marks_value} added successfully.\n")

# After the loop ends, print how many new subjects were added and the updated average across all subjects (original + new).

total = sum(marks)
average = round(total / len(marks), 2)

print(f"\nNew subjects added: {new_subjects_count}")
print(f"Updated average across all subjects: {average}")

# Handle invalid marks inputs (non-numeric values, numbers outside 0–100) with a warning message — do not crash, and do not add invalid entries to the list.
# This is handled in the code above with checks for numeric input and range validation.


```

```python
# Task 3 - Class Performance Summary

class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]

# 1. Loop through class_data and for each student.
# 1.1 Compute their average (rounded to 2 decimal places).

for student in class_data:
    name = student[0]
    marks = student[1]
    average = round(sum(marks) / len(marks), 2)
    
# 1.2 Assign a result status: Pass if average ≥ 60, else Fail.
    result = "Pass" if average >= 60 else "Fail"

# 2. Print a formatted class report:

print(f"{'Name':<15} | {'Average':>7} | Status")
print("-" * 40)

for name, marks in class_data:
    average = round(sum(marks) / len(marks), 2)
    status = "Pass" if average >= 60 else "Fail"
    print(f"{name:<15} | {average:>7.2f} | {status}")

# 3. After the table, print:
# 3.1 Number of students who passed and who failed
pass_count = sum(1 for _, marks in class_data if sum(marks) / len(marks) >= 60)
fail_count = len(class_data) - pass_count
print(f"\nTotal Passed: {pass_count}")
print(f"Total Failed: {fail_count}")

# 3.2 Class topper (name + average)
topper = max(class_data, key=lambda x: sum(x[1]) / len(x[1]))
print(f"Class Topper: {topper[0]} with an average of {round(sum(topper[1]) / len(topper[1]), 2)}")

# 3.3 Class average (average of all five students' averages)

class_average = round(sum(sum(marks) / len(marks) for _, marks in class_data) / len(class_data), 2)
print(f"Class Average: {class_average}")

```

```python
# Task 4 - String Manipulation Utility
# Perform each of the following operations on the stripped version of the essay (apply step 1 first, then use that result for all subsequent steps). 
# Print the result of each step separately with a label.

essay = "  python is a versatile language. it supports object oriented, functional, and procedural programming. python is widely used in data science and machine learning.  "

# 1. Strip leading and trailing whitespace from essay. Store this as clean_essay and use it for all steps below.
clean_essay = essay.strip()
print(f"Cleaned Essay:\n{clean_essay}\n")

# 2. Convert clean_essay to Title Case and print it.
title_case_essay = clean_essay.title()
print(f"Title Case Essay:\n{title_case_essay}\n")

# 3. Count how many times the word "python" appears in clean_essay (case-insensitive). Print the count.
python_count = clean_essay.lower().count("python")
print(f"Count of 'python' (case-insensitive): {python_count}\n")

# 4. Replace every occurrence of "python" in clean_essay with "Python 🐍" and print the result.
replaced_essay = clean_essay.replace("python", "Python 🐍")
print(f"Replaced Essay:\n{replaced_essay}\n")

# 5. Split clean_essay into individual sentences by splitting on ". " (a period followed by a space). Print the resulting list.
sentences = clean_essay.split(". ")
print(f"Sentences:\n{sentences}\n")

# 6. Print each sentence on a new line, numbered starting from 1. Add a "." at the end of each sentence if it does not already end with one.
print("Formatted Sentences:")
for i, sentence in enumerate(sentences, start=1):
    sentence = sentence.strip()
    if not sentence.endswith("."):
        sentence += "."
    print(f"{i}. {sentence}")

```
